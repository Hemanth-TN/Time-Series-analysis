```python
import pandas as pd
import plotly.graph_objects as go
```

```python
air_india_df = pd.read_csv("aqli_country_data_India.csv")
air_india_df.head()
```

```python
air_india_df.set_index('name', drop=True, inplace=True)
```

```python

air_india_df.rename({'name0':'country'}, axis=1, inplace=True)
# air_india_df.rename_axis({'name': 'state'}, axis=0, inplace=True)
air_india_df.index.name = 'state'
air_india_df.head()
```

```python
type(air_india_df.columns)
```

```python
def get_columns(column_names:pd.core.indexes.base.Index, string:str):
    return [f for f in column_names if string in f]
```

```python
pm_columns = get_columns(air_india_df.columns, 'pm')
who_columns = get_columns(air_india_df.columns, 'who')
nat_columns = get_columns(air_india_df.columns, 'nat')
```

```python
pm_india_df = air_india_df[pm_columns].copy()
who_india_df = air_india_df[who_columns].copy()
nat_india_df = air_india_df[nat_columns].copy()
```

```python
pm_india_df.columns = [f[-4:] for f in pm_india_df.columns]
who_india_df.columns = [f[-4:] for f in pm_india_df.columns]
nat_india_df.columns = [f[-4:] for f in pm_india_df.columns]
```

```python
pm_india_df.head()
```

```python
nat_india_df.columns
```

```python
#converting the columns names from strings to integers
pm_india_df.columns = [int(c) for c in pm_india_df.columns]
who_india_df.columns = [int(c) for c in who_india_df.columns]
nat_india_df.columns = [int(c) for c in nat_india_df.columns]
```

```python
pm_india_df = pm_india_df.T
who_india_df = who_india_df.T
nat_india_df = nat_india_df.T
```

```python
(who_india_df > nat_india_df).sum()
```

```python
# pm_india_df.T.plot(backend='hvplot', height=500, width=1000,title='PM values in India')
```

```python
data=[go.Scatter(x=who_india_df.index,
                y=who_india_df[c],
                name=c,
                mode='markers+lines') for c in who_india_df.columns]
layout = go.Layout(title={'text': 'Based on WHO guidelines in India',
                          'x': 0.5},
                   xaxis={'title':'Year',
                          'color': 'black'},
                   yaxis={'title': 'Increase in Life Expectancy (Years)',
                          'color': 'black'})
fig = go.Figure(data=data, layout=layout)
fig.write_html('WHO_Values_in_India.html')
```

```python
data=[go.Scatter(x=nat_india_df.index,
                y=nat_india_df[c],
                name=c,
                mode='markers+lines') for c in nat_india_df.columns]
layout = go.Layout(title={'text': 'Based on National guidelines in India',
                          'x': 0.5},
                   xaxis={'title':'Year',
                          'color': 'black'},
                   yaxis={'title': 'Increase in Life Expectancy (Years)',
                          'color': 'black'})
fig = go.Figure(data=data, layout=layout)
fig.write_html('nat_Values_in_India.html')
```

```python
data=[go.Scatter(x=pm_india_df.index,
                y=pm_india_df[c],
                name=c,
                mode='markers+lines') for c in pm_india_df.columns]
layout = go.Layout(title={'text': 'PM2.5 in USA',
                          'x': 0.5},
                   xaxis={'title':'Year',
                          'color': 'black'},
                   yaxis={'title': 'Micrograms/m3',
                          'color': 'black'})
fig = go.Figure(data=data, layout=layout)
fig.write_html('PM_Values_in_India.html')
```
