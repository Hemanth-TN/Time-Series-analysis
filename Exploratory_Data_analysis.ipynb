## Exploratory data Analysis

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_cumulative_data(place, particle):
    files = [f for f in os.listdir("./Data") if f"{place}_{particle}" in f]
    i=0
    cum_df = pd.DataFrame()
    for file in files:
        df = pd.read_csv(f"./Data/{file}", index_col=0, parse_dates=True)
        cum_df = pd.concat([cum_df, df], axis=0)
    
    return cum_df
```

```python
mojave_o3_df = get_cumulative_data("mojave", "o3")
mojave_pm25_data = get_cumulative_data("mojave", "pm2.5")
mojave_pm10_data = get_cumulative_data("mojave", "pm10")
```

## Visual exploratory data analsysis

```python
mojave_o3_df.plot(ylabel='O3 particle (ppm)', figsize=(10,5));
```

```python
mojave_pm25_data.plot(ylabel=r'PM 2.5 particle ($\mu g/m^3$)', figsize=(10,5));
```

```python
mojave_pm10_data.plot(ylabel=r'Pm 10 particle ($\mu g/m^3$)', figsize=(10,5));
```

```python
import matplotlib.pyplot as plt
import numpy as np
def get_histogram_box(df, limits=None):
    if limits is not None:
        df = df.loc[np.logical_and(df.iloc[:,0]>limits[0] ,df.iloc[:,0]<limits[1])]

    ax = df.plot(kind='hist', density=True, alpha=0.5)
    df.plot(kind='box', secondary_y=True, vert=False, ax=ax, color='r')
    plt.title(df.columns[0])
    plt.show()
```

```python
get_histogram_box(mojave_o3_df)
```

```python
get_histogram_box(mojave_pm25_data)
```

```python
get_histogram_box(mojave_pm10_data)
```

```python
get_histogram_box(mojave_pm10_data, limits=(0,100))
```

```python
get_histogram_box(mojave_pm25_data, limits=(0,50))
```
